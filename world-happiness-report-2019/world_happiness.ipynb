```python
import pandas as pd 
import pandas_profiling
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data =  pd.read_csv('world-happiness-report-2019.csv')
```

```python
data.head()
```

```python
pandas_profiling.ProfileReport(data)
```

```python
data = data.dropna()
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
plt.scatter(data['Ladder'], data['Social_support'])
plt.xlabel('Ladder')
plt.ylabel('Social support')
plt.title('Correlation between Ladder and social support')
plt.show()
```

```python
plt.scatter(data['Ladder'], data['Log_of_GDP\nper_capita'])
plt.xlabel('Ladder')
plt.ylabel('Log of GDP per capita')
plt.title('Correlation between Ladder and Log of GDP per capita')
plt.show()
```

```python
plt.scatter(data['Ladder'], data['Healthy_life\nexpectancy'])
plt.xlabel('Ladder')
plt.ylabel('Healthy life expectancy')
plt.title('Correlation between Ladder and Healthy life expectancy')
plt.show()
```
